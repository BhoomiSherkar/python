```python
class person:
 def __init__(self,name,age):
   self.name = name
   self.age = age

 def intro(self):
  print("My name is",self.name,"and I am", self.age, "years old.")

person1 = person("Bhoomi", 19)
person1.intro()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p1 = Person("John", 36)
p1.myfunc()
```
