```python
import numpy as np

# Create a 2D array manually
data = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print("The array is:\n", data)
print("Data type:", type(data))
print("Number of dimensions:", np.ndim(data))
print("Total elements:", data.size)
print("Shape (rows, columns):", data.shape)
```

```python
# Access an element
element = data[1, 2]  # Row 1, Column 2 (value: 7)
print("Element at (1, 2):", element)

# Slice a row
row_slice = data[0, :]  # First row
print("First row:", row_slice)

# Slice a column
column_slice = data[:, 1]  # Second column
print("Second column:", column_slice)
```

```python
import numpy as np

# Create the arrays A and B
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8, 9], [10, 11, 12]])

# Add the arrays element-wise
result = A + B

print("Array A:\n", A)
print("Array B:\n", B)
print("Sum of A and B:\n", result)
```

```python
import numpy as np

# Create two 2D arrays
x1 = np.array([[1, 2], [0, 5]])
x2 = np.array([[3, 1], [7, 1]])

# Multiply x1 and x2 element-wise
result = np.multiply(x1, x2)

print("Resultant output array:\n", result)
```
